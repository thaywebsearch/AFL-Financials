```python
# Criar uma tabela baseada em tabulate com empresas do meu Portfólio no setor das energias renováveis

from tabulate import tabulate

mydata = [
    ["EDP Renováveis", "Energias renováveis", "País"]
]

head = ["Empresa", "Indústria", "País"]

print(tabulate(mydata, headers=head, tablefmt="grid"))
```
